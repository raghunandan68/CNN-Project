```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dataset_path=r"D:\Tekworks\Deep-Learning(DL)\CNN\Cat_Dog_Dataset\PetImages"
```

```python
datagen=ImageDataGenerator(
    # Normalize pixels
    rescale=1./255,
    #Data augmentation
    shear_range=0.2,zoom_range=0.2,horizontal_flip=True,
    #Split dataset for training and validation sets
    validation_split=0.2)
```

```python
#Load training data
train_data=datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
```

```python
val_data=datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
print(train_data.class_indices)
```

```python
# Building CNN
model=keras.models.Sequential()
```

```python
# First convolution layer
model.add(
    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(128,128,3)
    )
)
```

```python
# First pooling layer
model.add(
    layers.MaxPooling2D(
        pool_size=(2,2),
    )
)
```

```python
# Second Convolution layer
model.add(
    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu'
    )
)
```

```python
# Second pooling layer
model.add(
    layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
# Third Convolution layer
model.add(
    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu'
    )
)
```

```python
# Third Pooling layer
model.add(
    layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
# Flattening layer
model.add(layers.Flatten())
```

```python
# Dense layer
model.add(layers.Dense(128,activation='relu'))
```

```python
# Output layer
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
# Model architecture
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(train_data,validation_data=val_data,epochs=2)
```

```python
loss,acc=model.evaluate(val_data)
print(f"Accuracy : {acc}")
print(f"Loss : {loss}")
```

```python
pred=model.predict(val_data)
pred_classes=(pred>0.5).astype(int)
image,label=next(train_data)
print("Original labels : ")
print(label[:5])
print("Predicted labels : ")
print(pred_classes[:5])
```
